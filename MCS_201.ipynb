```python
#Map function applies a given function to each element of an iterable like list,tuple etc
def square(n):
 return n * n
#list of numbers
numbers = [1,2,3,4,5]
#using map function
squares = list(map(square, numbers))
print("squares of numbers : ",squares)
```

```python
#Lambda function to calculate cube
#lambda argument : expression
cube = lambda x : x **3
#enter the number
num = int(input("Enter a number : "))
#Display the result
print("The cube of the number is :", cube(num))
```

```python
#Python code to create database and Connect
import mysql.connector

#Assuming MYSQL server is running locally
host ="localhost"
username ="your_username" #replace with your MYSQL username
password ="your_password" #replace with your MYSQL username

#connect to MYSQL server(no database required)
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password
)

#Create  a cursor object to execute SQL queries
cursor=connection.cursor()

#Create a new database  named 'student_DB'
database_name = "student_DB"
create_db_query = f"CREATE DATABASE IF NOT EXISTS {database_name}"
cursor.execute(create_db_query)

print(f"Database '{database_name}'created successfully.")

#Connect to the 'student_DB' database

connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password
    database=database_name
)
print(f"Connected to database'{database_name}'")

#Close the cursor and connection
cursor.close()
connection.close()
```

```python
#Example for overriding
class Animal:
  def speak(self):
    return "I make some sound"
class Dog(Animal):
  def speak(self):
    return "Bow!"
#Create objects
animal = Animal()
dog = Dog()
#Call the speak method
print(f"Animal says : {animal.speak()}")
print(f"Dog says : {dog.speak()}")
```
