```python
#Python Practice Problems – (Variables,Operators, Strings, Lists, Tuples)

#1. A financial analyst stores monthly revenues in a list: [120000, 135000, 128000,142000, 150000].
#Find the difference between the highest and lowest revenue without using loops.

Revenue_list = [120000, 135000, 128000,142000, 150000]
Revenue_list.sort()
Difference = Revenue_list[-1]-Revenue_list[0]
print(Difference)
```

```python
#2. Given the tuple account = ('Asha', 'Savings', 56000, 'KCH'), unpack the values into
#separate variables and print a formatted string: 'Asha has ₹56000 in her Savings account (Branch: KCH)'.

account = ('Asha', 'Savings', 56000, 'KCH')
Name = account[0]
Acc_type = account[1]
Balance_amount = account[2]
Branch_name = account[3]
print(f'{Name} has ₹{Balance_amount} in her {Acc_type} account (Branch: {Branch_name})')
```

```python
#3. Convert the list of string numbers ['45000', '56000', '38000'] to integers and find their average balance

List_of_numbers = ['45000', '56000', '38000']
New_list = [int(i) for i in List_of_numbers]
Count = len(New_list)
Total_sum = New_list[0]+New_list[1]+New_list[2]
Average = Total_sum/Count
print(Average)
#New_list = list(map(int, List_of_numbers))
```

```python
#4. Create a list with elements: ['Kochi', 'Chennai', 'Mumbai', 'Delhi', 'Pune']. Replace
#'Delhi' with 'Hyderabad' and print the new list in reverse order using slicing

City_list = ['Kochi', 'Chennai', 'Mumbai', 'Delhi', 'Pune']
City_list[3] = 'Hyderabad'
reverse_city_list = City_list[::-1]
print(reverse_city_list)
```

```python
#5. You have two tuples: interest_2024 = (7.1, 7.3, 7.2) and interest_2025 = (7.4, 7.5).
#Combine them and print the average of all interest rates.

t1=(7.1,7.3,7.2)
t2=(7.4,7.5)
t3=sum(t1+t2)/len(t1+t2)
print(f"The average of all interest rates is: {t3:.2f}")
```

```python
#9.A tuple customer = ('Kiran', 'Loan', 80000, 7.5). Calculate and print the total amount
#payable after 1 year if interest is simple interest based on rate 7.5%.

tuple_cust = ('Kiran', 'Loan', 80000, 7.5)
principal = tuple_cust[2]
interest_rate = tuple_cust[3]
time = 1
interest = principal * (interest_rate / 100) * time
total_amount = principal + interest
print(f"The total amount payable after one year is RS:{total_amount:.2f}")
```
