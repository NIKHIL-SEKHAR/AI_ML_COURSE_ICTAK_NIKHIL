```python
sales = np.array([1000, 2000, 3000, 4000, 5000])
costs = np.array([500, 1000, 1500, 2000, 2500])

profit = sales - costs
profit
```

```python
profit_percentage = (profit / sales) * 100
PRO_PER = profit_percentage.round(2)
PRO_PER
```



7.Given stock prices, compute daily returns using vectorized NumPy operations.


```python
prices = np.array([100, 102, 98, 105, 107, 103, 110])
daily_returns = (prices[1:] - prices[:-1]) / prices[:-1]
print("Stock Prices:", prices)
print("Daily Returns:", daily_returns.round(4))
```



8.Convert an array of annual interest rates into monthly rates using vectorization.


```python
annual_rates = np.array([0.06, 0.08, 0.12, 0.05])
monthly_rates = annual_rates / 12
print("Annual Interest Rates:", annual_rates)
print("Monthly Interest Rates:", monthly_rates.round(4))
```



9.Normalize a given array of portfolio weights so that the sum becomes 1.


```python
weights = np.array([30, 25, 20, 15, 10])
normalized_weights = weights / weights.sum()
print("Original weights:", weights)
print("Normalized (sum=1):", normalized_weights.round(4))
```

Create series and stats.

```python
import pandas as pd
data = [1200, 1500, 1800, 2000, 2200]
s = pd.Series(data, index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])
print("Series:\n", s)
print("Mean:", s.mean())
print("Max:", s.max())
print("Index of max:", s.idxmax())  # or s.index[s.argmax()]
```

```python
df = pd.DataFrame({
    "Account": [2001, 2002, 2003, 2004],
    "Balance": [6000, 14000, 8000, 13000]
})
print("Original DF:\n", df)
print("Balance > 7000:\n", df[df['Balance'] > 7000])
print("Balance between 7000 and 14000:\n",
      df[df['Balance'].between(7000, 13000)])
```
